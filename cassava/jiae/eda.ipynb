```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
data_path='/kaggle/input/competitions/cassava-leaf-disease-classification'
```

```python
df = pd.read_csv(os.path.join(data_path, "train.csv"))
```

```python
# 클래스 분포 시각화
label_counts = df['label'].value_counts()

print(label_counts)

plt.figure()
label_counts.sort_index().plot(kind='bar')

plt.title("Class Distribution")
plt.xlabel("Label")
plt.ylabel("Count")

plt.show()
```

```python
# 샘플 이미지 시각화
image_dir = os.path.join(data_path, "train_images")

labels = df['label'].unique()

plt.figure(figsize=(10, 10))

idx = 1

for label in labels:
    sample = df[df['label'] == label].sample(2)

    for _, row in sample.iterrows():
        img_path = os.path.join(image_dir, row['image_id'])
        img = Image.open(img_path)

        plt.subplot(len(labels), 2, idx)
        plt.imshow(img)
        plt.title(f"Label: {label}")
        plt.axis('off')

        idx += 1

plt.tight_layout()
plt.show()
```
